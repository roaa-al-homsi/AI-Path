## 1. Introduction to NumPy

```python

import time

lst = list(range(1_000_000))
start = time.time()
[x**2 for x in lst]
print("Python list time:", time.time() - start)

arr = np.arange(1_000_000)
start = time.time()
arr ** 2
print("NumPy array time:", time.time() - start)
```

## 2. Creating and Inspecting Arrays

```python

a = np.array([1, 2, 3, 4])
print("Array:", a)
print("Shape:", a.shape)
print("Data type:", a.dtype)
print("Dimensions:", a.ndim)

print("\nZeros:\n", np.zeros((2,3)))
print("\nOnes:\n", np.ones((3,3)))
print("\nIdentity matrix:\n", np.eye(4))
```

```python

b = np.arange(10)
print("Original:", b)
print("Reshaped (2x5):\n", b.reshape(2,5))
print("Linspace:", np.linspace(0, 1, 5))
```

## 3. Array Operations and Broadcasting

```python

x = np.array([1, 2, 3])
y = np.array([10, 20, 30])
print("Addition:", x + y)
print("Multiplication:", x * y)
print("Dot Product:", np.dot(x, y))
print("Mean of y:", y.mean())

A = np.ones((3,3))
b = np.array([1,2,3])
print("\nBroadcasted addition:\n", A + b)
```

## 4. Aggregations and Logical Operations

```python

data = np.random.randint(0, 100, size=(5,5))
print("Data:\n", data)

print("Sum:", data.sum())
print("Mean (axis=0):", data.mean(axis=0))
print("Max (axis=1):", data.max(axis=1))

print("\nValues > 50:\n", data[data > 50])

#axis=0 ==> apply on columns.
#axis=1 ==>apply on rows.
```

## 5. Random Number Generation

```python

import numpy as np
np.random.seed(42)
print("Uniform:", np.random.rand(5))
print("Normal distribution:", np.random.randn(5))
print("Integers:", np.random.randint(0, 10, 5))

print(f"randint{np.random.randint(20,50,size=(2,3))}")
print(f"\n rand \n {np.random.rand(3,2)}")#values between 0 and 1 (positive numbers)
print(f"\nrandn \n {np.random.randn(3,2)}")#values between 0 and 1 using normal distribution (pos and neg numbers)
print(f"\n random \n {np.random.random((3,2))}")#like rand values between 0 and 1 (positive numbers) but with one argument
print(f"\n random \n {np.random.choice([1,1,2,4],size=(2,3))}")#choice random elements from collection i give it .
print(f"\n normal \n {np.random.normal(50, 10,size=3)}")#mean,std,size
```

## 6. ðŸŽ² Mini Project: Random Data Simulation & Analysis

```python

np.random.seed(123)
n_students = 500
ages = np.random.randint(18, 30, n_students)
scores = np.random.normal(70, 10, n_students).clip(0, 100)
study_hours = np.random.normal(5, 2, n_students).clip(0)

data = np.column_stack((ages, scores, study_hours))
print("Shape:", data.shape)
print("First 5 rows:\n", data[:5])
```

```python

print("Average score:", np.mean(scores))
print("Max study hours:", np.max(study_hours))
print("Age range:", np.min(ages), "-", np.max(ages))

corr = np.corrcoef(scores, study_hours)[0,1]
print("Correlation between score and study hours:", round(corr, 3))
```

```python

plt.figure(figsize=(6,4))
plt.scatter(study_hours, scores, alpha=1)#dot graph representation
plt.title("Scores vs Study Hours")
plt.xlabel("Study Hours per Day")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
scores = np.random.randint(0, 100, size=100)
print(scores)
print("shape:", scores.shape)

num_above_75 = len(scores[scores > 75])
print("Number of students scoring above 75:", num_above_75)

top_10_percent_cutoff = np.percentile(scores, 90)
print("\nTop 10% cutoff score:", top_10_percent_cutoff)

sleep_hours = np.random.normal(7, 1, 100)
data = np.column_stack((scores, sleep_hours))

#  correlation
corr_sleep_score = np.corrcoef(sleep_hours, scores)[0,1]
print("Correlation between sleep hours and scores:", round(corr_sleep_score,3))

#  scatter
plt.figure(figsize=(6,4))
plt.scatter(sleep_hours, scores, alpha=0.7)
plt.title("Sleep Hours vs Scores")
plt.xlabel("Sleep Hours")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```
